```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as lin
```

##**Assignment 3**
* 행렬 방정식 풀기  
  * 다음 행렬 방적식을 'Pseudo inverse matrix'를 이용해 풀어보자
  * $\mathbf{A}^\intercal \mathbf{A}$의 역행렬이 존재한다고 가정  
$$
\begin{aligned}  
Ax = B\\  
A =  \left[
\begin{matrix}
    0 & 1 \\
    1 & 1 \\
    2 & 1 \\
    3 & 1 \\
\end{matrix}
\right]\\
B = \left[
\begin{matrix}
    -1 \\
    0.2 \\
    0.9 \\
    2.1 \\
\end{matrix}
\right]
\end{aligned}
$$

```python
#3
matA = np.array([[0,1],[1,1],[2,1],[3,1]])
matB = np.array([[-1], [0.2],[0.9],[2.1]])
A = lin.pinv(matA) #matA가 정방 행렬이 아니므로 유사 역행렬 사용
print(A) #matA의 유사 역행렬
print("-----------------")
print(np.matmul(A,matB)) #X
```

##**Assignment 4**
* 2x2 그래프 그리기
  * $y = x$
  * $y = x^2$
  * $y = sin(x)$
  * $y = cos(x)$

```python
#4
x = np.arange(0, 3 * np.pi, 0.1)
y = x
y1 = x**2 
y_sin = np.sin(x)
y_cos = np.cos(x)


# 그래프 간격 조절
left = 0.0
bottom = 0.0
right = 2.0
top = 2.0
wspace = 0.5
hspace = 0.5
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Y = X')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title('Y = X^2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(2,2,3)
plt.plot(x,y_sin)
plt.title('Sine')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(2,2,4)
plt.plot(x,y_cos)
plt.title('Cosine')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
